```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Use seaborn style
sns.set_style("whitegrid")

# 1. Load CSV files (skip first 2 rows with headers)
aapl = pd.read_csv("AAPL.csv", index_col="Date", parse_dates=True, skiprows=2)
tsla = pd.read_csv("TSLA.csv", index_col="Date", parse_dates=True, skiprows=2)
msft = pd.read_csv("MSFT.csv", index_col="Date", parse_dates=True, skiprows=2)

# 2. Rename columns properly
aapl.columns = ["Close", "High", "Low", "Open", "Volume"]
tsla.columns = ["Close", "High", "Low", "Open", "Volume"]
msft.columns = ["Close", "High", "Low", "Open", "Volume"]

# 3. Compute daily returns
returns = pd.DataFrame({
    "AAPL": aapl["Close"].pct_change(),
    "TSLA": tsla["Close"].pct_change(),
    "MSFT": msft["Close"].pct_change()
}).dropna()

# 4. Covariance & correlation matrices
cov_matrix = returns.cov()
corr_matrix = returns.corr()

print("Covariance matrix:\n", cov_matrix, "\n")
print("Correlation matrix:\n", corr_matrix, "\n")

# 5. Monte Carlo simulation for portfolio optimization
np.random.seed(42)
n_portfolios = 5000
results = []

for _ in range(n_portfolios):
    weights = np.random.random(3)
    weights /= np.sum(weights)  # normalize
    
    portfolio_return = np.sum(weights * returns.mean()) * 252
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix * 252, weights)))
    sharpe_ratio = portfolio_return / portfolio_volatility
    
    results.append([portfolio_return, portfolio_volatility, sharpe_ratio, weights])

results_df = pd.DataFrame(results, columns=["Return", "Volatility", "Sharpe", "Weights"])

# 6. Find portfolios of interest
min_vol_port = results_df.iloc[results_df["Volatility"].idxmin()]
max_sharpe_port = results_df.iloc[results_df["Sharpe"].idxmax()]

# Function to pretty-print portfolio info
def print_portfolio(name, port):
    weights = [float(w) for w in port["Weights"]]  # cast to Python float
    weights_dict = {
        "AAPL": round(weights[0]*100, 2),
        "TSLA": round(weights[1]*100, 2),
        "MSFT": round(weights[2]*100, 2)
    }
    print(f"\n{name} Portfolio:")
    print(f"Expected Annual Return: {port['Return']*100:.2f}%")
    print(f"Annual Volatility (Risk): {port['Volatility']*100:.2f}%")
    print(f"Sharpe Ratio: {port['Sharpe']:.2f}")
    print("Weights (%):", weights_dict)
    return weights_dict

min_weights = print_portfolio("Minimum Volatility", min_vol_port)
max_weights = print_portfolio("Maximum Sharpe", max_sharpe_port)

# 7. Plot Efficient Frontier
plt.figure(figsize=(10,6))
plt.scatter(results_df["Volatility"], results_df["Return"],
            c=results_df["Sharpe"], cmap="viridis", alpha=0.6)
plt.colorbar(label="Sharpe Ratio")

# Plot stars
plt.scatter(float(min_vol_port["Volatility"]), float(min_vol_port["Return"]),
            color="red", marker="*", s=200, label="Min Volatility", zorder=5)
plt.scatter(float(max_sharpe_port["Volatility"]), float(max_sharpe_port["Return"]),
            color="yellow", marker="*", s=200, edgecolor="black", label="Max Sharpe", zorder=6)

# Annotate weights near stars
plt.annotate(f"Weights:\n{min_weights}",
             (float(min_vol_port["Volatility"]), float(min_vol_port["Return"])),
             xytext=(20, -30), textcoords="offset points",
             arrowprops=dict(arrowstyle="->", color="red"), fontsize=9, color="red")

plt.annotate(f"Weights:\n{max_weights}",
             (float(max_sharpe_port["Volatility"]), float(max_sharpe_port["Return"])),
             xytext=(20, 30), textcoords="offset points",
             arrowprops=dict(arrowstyle="->", color="black"), fontsize=9, color="black")

plt.xlabel("Volatility (Risk)")
plt.ylabel("Expected Return")
plt.title("Efficient Frontier (Markowitz Portfolio Theory)")
plt.legend()

plt.show()
```

```python
# 8. Save outputs
import os

# Create folders if not exist
os.makedirs("images", exist_ok=True)
os.makedirs("results", exist_ok=True)

# Save efficient frontier plot
plt.figure(figsize=(10,6))
plt.scatter(results_df["Volatility"], results_df["Return"],
            c=results_df["Sharpe"], cmap="viridis", alpha=0.6)
plt.colorbar(label="Sharpe Ratio")

# Stars
plt.scatter(float(min_vol_port["Volatility"]), float(min_vol_port["Return"]),
            color="red", marker="*", s=200, label="Min Volatility", zorder=5)
plt.scatter(float(max_sharpe_port["Volatility"]), float(max_sharpe_port["Return"]),
            color="yellow", marker="*", s=200, edgecolor="black", label="Max Sharpe", zorder=6)

plt.xlabel("Volatility (Risk)")
plt.ylabel("Expected Return")
plt.title("Efficient Frontier (Markowitz Portfolio Theory)")
plt.legend()

plt.savefig("images/efficient_frontier.png", dpi=300, bbox_inches="tight")
plt.close()

# Save portfolio stats
summary = pd.DataFrame({
    "Portfolio": ["Minimum Volatility", "Maximum Sharpe"],
    "Expected Return (%)": [round(min_vol_port["Return"]*100, 2),
                            round(max_sharpe_port["Return"]*100, 2)],
    "Volatility (%)": [round(min_vol_port["Volatility"]*100, 2),
                       round(max_sharpe_port["Volatility"]*100, 2)],
    "Sharpe Ratio": [round(min_vol_port["Sharpe"], 2),
                     round(max_sharpe_port["Sharpe"], 2)],
    "Weights": [min_weights, max_weights]
})

summary.to_csv("results/portfolio_summary.csv", index=False)
print("\nResults saved in 'results/portfolio_summary.csv' and 'images/efficient_frontier.png'")
```
